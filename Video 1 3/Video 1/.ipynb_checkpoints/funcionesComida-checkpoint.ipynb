```python
from manim import *
import networkx as nx
import numpy as np
import math
```

```python
# Restaurante
def restaurante(self):
        r = SVGMobject("Food/restaurante.svg")
        r.scale(3.3)
        restaurant = Text('IRISFERA', color=BLACK)
        restaurant.move_to([-0.1, 1.5, 0]).scale(1.1)
        self.play(Write(r), run_time=2.5)
        self.play(Write(restaurant))
        self.play(restaurant.animate.move_to([-13, 17, 0]).scale(6), 
                  FadeOut(r, target_position= [-12, 5, 0], scale=7.5), 
                  run_time=1.3)
```

```python
def tipos_pasta(self):
    # Crea el primer bloque de texto "Tipos de pasta"
    tipos1 = Tex("3 tipos")
    pasta = Tex("de pasta")

    #Agregar tipos de pasta
    espagueti = SVGMobject("Food/espagueti.svg")
    espagueti.move_to((np.array([-5, 2, 0]))).scale(1.5)


    fusili = SVGMobject("Food/fusili.svg")
    fusili.move_to((np.array([-2, 2, 0]))).scale(1)

    macaroni = SVGMobject("Food/macaroni.svg")
    macaroni.move_to((np.array([1, 2, 0]))).scale(1)


    # Define las posiciones de las líneas
    tipos1.next_to(ORIGIN, LEFT)
    pasta.next_to(tipos1, RIGHT)
    self.play(Write(tipos1))
    self.play(Write(pasta))

    self.play(
        tipos1.animate.move_to([5, 2.5, 0]),
        pasta.animate.next_to([5, 2.5, 0], DOWN),
        Write(espagueti), run_time = .4
    )

    self.play(Write(fusili), run_time=.4)
    self.play(Write(macaroni), run_time=.4)
    
    return (tipos1,pasta,espagueti,fusili,macaroni)
```

```python
def tipos_pasta_2(self):
    # Crea el primer bloque de texto "Tipos de pasta"
    tipos1 = Tex("2 tipos")
    pasta = Tex("de pasta")

    #Agregar tipos de pasta
    espagueti = SVGMobject("Food/espagueti.svg")
    espagueti.move_to((np.array([-5, 2, 0]))).scale(.95).rotate(-.3)


    fusili = SVGMobject("Food/fusili.svg")
    fusili.move_to((np.array([-3, 2, 0]))).scale(.68)


    # Define las posiciones de las líneas
    tipos1.next_to(ORIGIN, LEFT)
    pasta.next_to(tipos1, RIGHT)
    self.play(Write(tipos1))
    self.play(Write(pasta))

    
    self.play(
        tipos1.animate.move_to([-5, 3.5, 0]),
        pasta.animate.move_to([-3.2, 3.5, 0]),
        Write(espagueti), run_time = .4
    )
    self.play(Write(fusili), run_time=.4)
    return (tipos1,pasta,espagueti,fusili)
```

```python
def tipos_salsa(self):
    # Crea el segundo bloque de texto "Tipos de salsa"
    tipos2 = Tex("3 tipos")
    salsa = Tex("de salsa")

    # Tipos de salsa
    salsaPom = SVGMobject("Food/SalsaPomodoro.svg")
    salsaPom.move_to((np.array([-5, -2, 0]))).scale(1.5)

    salsaPar = SVGMobject("Food/SalsaParmesano.svg")
    salsaPar.move_to((np.array([-2, -2, 0]))).scale(1.5)


    salsaAjo = SVGMobject("Food/SalsaAjo.svg")
    salsaAjo.move_to((np.array([1, -2, 0]))).scale(1.5)

    # Define las posiciones de las líneas
    tipos2.next_to(ORIGIN, LEFT)
    salsa.next_to(tipos2, RIGHT)
    self.play(Write(tipos2))
    self.play(Write(salsa))

    self.play(
        salsa.animate.move_to([5, -1.5, 0]),
        tipos2.animate.next_to([5, -1.5, 0], UP),
        Write(salsaPom), run_time = .6
    )

    self.play(Write(salsaPar), run_time=.6)
    self.play(Write(salsaAjo), run_time=1)
    
    return (tipos2,salsa, salsaPom, salsaPar, salsaAjo)
```

```python
def tipos_salsa_2(self):
    # Crea el segundo bloque de texto "Tipos de salsa"
    tipos2 = Tex("4 tipos")
    salsa = Tex("de salsa")

    # Tipos de salsa
    salsaPom = SVGMobject("Food/SalsaPomodoro.svg")
    salsaPom.move_to((np.array([-5.5, -.5, 0]))).scale(.9)

    salsaPar = SVGMobject("Food/SalsaParmesano.svg")
    salsaPar.move_to((np.array([-4, -.5, 0]))).scale(.9)

    
    salsaPes = SVGMobject("Food/SalsaPesto.svg")
    salsaPes.move_to((np.array([-2.5, -.5, 0]))).scale(.9)

    salsaCham = SVGMobject("Food/SalsaChampiñon.svg")
    salsaCham.move_to((np.array([-1, -.5, 0]))).scale(.9)

    # Define las posiciones de las líneas
    tipos2.next_to(ORIGIN, LEFT)
    salsa.next_to(tipos2, RIGHT)
    self.play(Write(tipos2))
    self.play(Write(salsa))

    self.play(
        salsa.animate.move_to([-3.2, .8, 0]),
        tipos2.animate.next_to([-6, .8, 0]),
        Write(salsaPom), run_time = .8
    )


    self.play(Write(salsaPar), run_time=.8)
    #self.play(Write(salsaAjo), run_time=1)

    self.play(Write(salsaPes), run_time=.8)
    self.play(Write(salsaCham), run_time=1)
    return (tipos2,salsa, salsaPom, salsaPar, salsaPes,salsaCham)
```

```python
def tipos_proteina(self):
    # Crea el primer bloque de texto "Tipos de pasta"
    tipos = Tex("2 tipos").move_to([2, 3.5, 0])
    proteina = Tex("de proteina")
    proteina.next_to(tipos, RIGHT)
    
    #Agregar tipos de pasta
    salmon = SVGMobject("Food/salmon.svg")
    salmon.move_to((np.array([-5, -3, 0]))).scale(.5)


    carne = SVGMobject("Food/carne.svg")
    carne.move_to((np.array([-2, -3, 0]))).scale(.5)

    
    self.play(Write(tipos))
    self.play(Write(proteina))

    self.play(
        tipos.animate.move_to([-5, -1.8, 0]),
        #proteina.animate.move_to([-2.9, -1.8, 0]),
        proteina.animate.next_to([-4.2, -1.8, 0], RIGHT),
        Write(salmon), run_time = .7
    )
    


    self.play(Write(carne), run_time=.6)
    
    return (tipos,proteina,salmon,carne)
```
