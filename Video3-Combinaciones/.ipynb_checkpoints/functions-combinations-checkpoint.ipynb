```python
from manim import VGroup
import networkx as nx
import numpy as np
import math
import copy
from itertools import permutations
```

```python
def crear_pentagono():
    pentagono = Polygon(
        [-4.2, -3.45, 0], [-4.2, -2.05, 0], [-2.8, -1.35, 0],
        [-1.3, -2.05, 0], [-1.3, -3.45, 0], color=PURE_RED, fill_opacity=0.5
    )
    pentagono.scale(1.25).move_to([-3.5, -2.3, 0])
    return pentagono

def crear_triangulo():
    triangulo = Polygon(
        [2.55, 0.2, 0], [4.3, 0.2, 0], [3.425, 3.7, 0], color=ORANGE, fill_opacity=0.5
    )
    return triangulo

def crear_circulo():
    circulo = Circle(radius=1.362, color=PINK, fill_opacity=0.5)
    circulo.move_to([3.5, -1.8, 0])
    return circulo

def crear_rectangulo():
    rectangulo = RoundedRectangle(corner_radius=0.9, height=2.65, width=4, color=BLUE_E, fill_opacity=0.5)
    return rectangulo
```

```python
def crear_permutaciones(atracciones):
    perm_atr = obtener_permutaciones(atracciones)
    list_comb, perm_list_vg, per_24 = procesar_permutaciones(perm_atr)
    return list_comb, perm_list_vg, perm_atr, per_24

def obtener_permutaciones(atracciones):
    perm_atr = []
    for i in range(len(atracciones)):
        sublist = atracciones[:i] + atracciones[i+1:]
        perm_atr.append(list(permutations(sublist)))
    return perm_atr

def procesar_permutaciones(perm_atr):
    x, y = -6.125, 3
    list_comb, perm_list_vg, per_24 = [], [], VGroup()
    i = 1

    for comb in perm_atr:
        comb_g = VGroup()    
        for perm in comb:
            perm_vg = VGroup()
            for at in perm:
                at_c = mover_atraccion(copy.deepcopy(at), x, y)
                perm_vg.add(at_c)
                x += 7 / 8
            
            x, y = ajustar_posiciones(x, y, i)
            perm_list_vg.append(perm_vg)
            comb_g.add(perm_vg)
            i += 1

        y = ajustar_y_inicial(y, i)
        list_comb.append(comb_g)
        per_24.add(comb_g)
        y -= 1
    
    return list_comb, perm_list_vg, per_24

def mover_atraccion(at, x, y):
    at.move_to([x, y, 0])
    return at

def ajustar_posiciones(x, y, i):
    if i % 2 == 1:
        x += 7 / 8
    if i % 2 == 0:
        y -= 1
        if i in [6, 8, 10, 18, 20, 22]:
            x = 0.875
        else:
            x = -6.125
    return x, y

def ajustar_y_inicial(y, i):
    if i == 7:
        return 4
    elif i == 13 or i == 19:
        return 0
    return y
```

```python
def crear_combinaciones(self, rectangulo, triangulo, pentagono, circulo):
    rect1 = RoundedRectangle(corner_radius=.9, height=2.65, width=4, color = BLUE_E, fill_opacity=.5)

    montaña_rusa = crear_copia_escalada(rect1, .2)
    carrusel = crear_copia_escalada(pentagono, .2)
    rueda_fortuna = crear_copia_escalada(triangulo, .2)
    castillo = crear_copia_escalada(circulo, .2)

    atracciones = [montaña_rusa, rueda_fortuna, carrusel, castillo]

    lista_comb, perm_lista_vg, perm_atr, per_24 = crear_permutaciones(atracciones)

    edge1 = Line(start=[-6.75, 0, 0], end=[6.75, 0, 0], z_index=1)
    edge2 = Line(start=[0, 3.65, 0], end=[0, -3.65, 0], z_index=1)

    self.play(
        Transform(rectangulo, lista_comb[0]),
        Transform(triangulo, lista_comb[1]),
        Transform(pentagono, lista_comb[2]),
        Transform(circulo, lista_comb[3]),
        Create(edge1),
        Create(edge2)
    )

    self.add(*lista_comb)
    self.remove(rectangulo, triangulo, pentagono, circulo)    
    self.wait(2.5)

    return lista_comb, perm_lista_vg

def crear_copia_escalada(figura, scale):
    copia_figura = copy.deepcopy(figura)
    copia_figura.scale(scale).set_fill(opacity=1)
    return copia_figura
```

```python
def crear_comb1(pentagono, triangulo, circulo):
    pentagono1 = crear_copia_escalada(pentagono, .6).move_to([-7/4, 2.5,0])
    triangulo1 = crear_copia_escalada(triangulo, .6).move_to([-7/2, 1.5, 0])
    circulo1 = crear_copia_escalada(circulo, .6).move_to([-21/4, 2.5,0])
    comb1 = VGroup(pentagono1, triangulo1, circulo1)
    return comb1

def crear_comb2(rectangulo, pentagono, circulo):
    rectangulo2 = crear_copia_escalada(rectangulo, .5).move_to([7/4, 2.5, 0])
    pentagono2 = crear_copia_escalada(pentagono, .6).move_to([7/2, 1, 0])
    circulo2 = crear_copia_escalada(circulo, .6).move_to([21/4, 2.5, 0])
    comb2 = VGroup(rectangulo2, pentagono2, circulo2)
    return comb2

def crear_comb3(triangulo, circulo, rectangulo):
    triangulo3 = crear_copia_escalada(triangulo, .6).move_to([-21/4, -2,0])
    circulo3 = crear_copia_escalada(circulo, .6).move_to([-7/2, -1.1, 0])
    rectangulo3 = crear_copia_escalada(rectangulo, .5).move_to([-7/4, -2.5, 0])
    comb3 = VGroup(triangulo3, circulo3, rectangulo3)
    return comb3

def crear_comb4(pentagono, rectangulo, triangulo):
    pentagono4 = crear_copia_escalada(pentagono, .6).move_to([7/4, -2.5, 0])
    rectangulo4 = crear_copia_escalada(rectangulo, .5).move_to([7/2, -1, 0])
    triangulo4 = crear_copia_escalada(triangulo, .6).move_to([21/4, -2.5, 0])
    comb4 = VGroup(pentagono4, rectangulo4, triangulo4)
    return comb4

```
